```python
import pandas as pd
import numpy as np
```

```python
pd.options.display.max_columns = 999
```

```python
df = pd.read_csv('../CapStone/LetsModelBaby.csv')

# dropping unecessary columns
df.drop('Unnamed: 0', inplace=True, axis=1)

df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.head()
```

```python
# SETTING THE DATE AS THE INDEX FOR MORE EFFICIENT EDA
df.set_index('Date', drop=False, inplace=True)
```

```python
# BATTING TEAM
df.groupby('BattingTeam1')['Total Runs in Game'].describe().sort_values('mean', ascending=False)
# THIS SHOWS THAT OVER THE COURSE OF THE ENTIRE DATAFRAME, WHICH ENCOMPASSES GAMES FROM MAY - OCTOBER IN 
# 2009 - 2016, THE MOST RUNS WERE SCORED IN COLORADO, BOSTON, TEXAS, ARIZONA, AND TORONTO'S BALLPARKS
```

```python
# WRITING CODE TO ITERATE OVER EVERY HOME STADIUM IN THE LEAGUE, AND PRINT OUT THE PERCENTAGES OF ALL THE GAMES
# PLAYED IN THAT STADIUM, HOW MANY HIT THE OVER AND HOW MANY HIT THE UNDER

for i in (set(df['BattingTeam1'].values)):
    print i
    print df.loc[df['BattingTeam1'] == i]['Over hit?'].value_counts(normalize=True)
    print "---------------------------"
```

```python
# SIMILAR TO ABOVE, WRITING CODE TO ITERATE OVER EVERY HOME STADIUM IN THE LEAGUE, AND PRINT OUT THE PERCENTAGES OF ALL THE GAMES
# PLAYED IN THAT STADIUM, HOW MANY TIMES THE HOME TEAM WON
for i in (set(df['BattingTeam1'].values)):
    print i
    print df.loc[df['BattingTeam1'] == i]['Home_Team_Win?'].value_counts(normalize=True)
    print "---------------------------"
```
